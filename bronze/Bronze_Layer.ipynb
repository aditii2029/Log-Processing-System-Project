****Read Posts JSON Data****

Loaded raw JSON data into a Spark DataFrame.

Used multiline option to correctly read multi-line JSON format.

```python
spark.sql("USE CATALOG log_catalog")
spark.sql("USE SCHEMA bronze")
```

****Add Ingestion Time (Posts)****

Added ingestion timestamp column to track when data was loaded.

Helps in data lineage and tracking.

```python
from pyspark.sql.functions import current_timestamp
```

```python
posts_raw = spark.read.format("json") \
    .option("multiline", "true") \
    .load("abfss://log-data@stlogprocessing123.dfs.core.windows.net/bronze/raw_logs/posts")
```

```python
posts_bronze = posts_raw.withColumn("ingestion_time",current_timestamp())
```

Saved data in Delta format to the Bronze layer storage.

Used overwrite mode to replace existing data.

```python
posts_bronze.write.format("delta") .mode("overwrite").save("abfss://log-data@stlogprocessing123.dfs.core.windows.net/bronze/posts")
```

****Read Comments JSON Data****

Loaded comments JSON data into DataFrame.

Multiline option ensures correct parsing of JSON.

```python
comments_raw = spark.read.format("json").option("multiline", "true").load("abfss://log-data@stlogprocessing123.dfs.core.windows.net/bronze/raw_logs/comments")
```

```python
display(comments_raw.limit(10))
```

***Add Ingestion Time***

Added ingestion timestamp to comments data.
Useful for tracking data load time.

```python
comments_bronze = comments_raw.withColumn("ingestion_time",current_timestamp())
```

Stored comments data in Delta format in Bronze layer.

Overwrite mode ensures latest data is stored.

```python
comments_bronze.write.format("delta").mode("overwrite").save("abfss://log-data@stlogprocessing123.dfs.core.windows.net/bronze/comments")
```

***SAVE AS TABLE***

Saved posts data as a table in the Bronze schema.
Allows querying using SQL.

```python
posts_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("log_catalog.bronze.posts")
```

***SAVE AS TABLE***

Saved comments data as a table in Bronze layer.
Makes data easily accessible for further processing.

```python
comments_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("log_catalog.bronze.comments")
```
